# Ekstraksi fitur TFID PTA Manajemen

```python
import pandas as pd

# Load data from pta_manajemen.csv
df = pd.read_csv('hasil_preprocessing_manajemen.csv')
```

```python
# Display basic information about the dataset
print("Dataset shape:", df.shape)
df
```

##  Ekstraksi Fitur TF-IDF

```python
import sys
!{sys.executable} -m pip install scikit-learn
```

```python
# Kolom teks hasil preprocessing
texts = df["hasil_preprocessing"].astype(str)

# Kalau ada kolom label
labels = df["abstrak_id"] 
```

##  TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()

# Fit & transform ke matriks TF-IDF
tfidf_matrix = vectorizer.fit_transform(texts)

# Konversi ke DataFrame agar lebih mudah dibaca
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)

# Tambahkan kolom label
tfidf_df["abstrak_id"] = labels.values
```

```python
print("Ukuran matriks TF-IDF:", tfidf_df.shape)
tfidf_df
```

```python
import sys
!{sys.executable} -m pip install openpyxl
```

```python
# Ganti nama file output menjadi .csv
output_file = "hasil_tfidf_pta_manajemen.csv"

# Gunakan metode .to_csv() untuk menyimpan ke file CSV
# index=False agar tidak ikut menyimpan indeks DataFrame
tfidf_df.to_csv(output_file, index=False)

print(f"Hasil TF-IDF berhasil disimpan ke file: {output_file}")
```
